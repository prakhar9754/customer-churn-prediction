# Customer Segmentation & Retention Analysis
## Phase 1: Business Understanding & Problem Framing
Olist E-commerce Dataset


## 1. Business Context

Olist is a Brazilian e-commerce platform connecting sellers to marketplaces.
The platform manages customers, sellers, payments, orders, and reviews.

In a competitive e-commerce landscape, customer retention and segmentation
are critical for increasing revenue and optimizing marketing strategies.


## 2. Problem Statement

Olist currently does not have a structured approach to:

- Identifying high-value customers
- Segmenting customers based on behavior
- Predicting customers at risk of churn
- Personalizing marketing strategies

This results in inefficient targeting and lower repeat purchases.


## 3. Business Objectives

1. Identify high-value and repeat customers.
2. Segment customers based on purchasing behavior.
3. Improve marketing targeting.
4. Predict customers at risk of churn.
5. Increase overall revenue and retention rate.


## 4. Analytical Objectives

This project includes two key data science problems:

### A. Customer Segmentation
- Type: Unsupervised Learning
- Method: RFM Analysis + Clustering (K-Means)

### B. Customer Retention / Churn Prediction
- Type: Supervised Learning (Binary Classification)
- Output: Churn (1 = churn, 0 = active)


Churn will be defined based on inactivity period.
Customers with no purchase activity within a defined window
will be labeled as churned.


## 5. Key Business KPIs

- Repeat Purchase Rate
- Customer Lifetime Value (CLV)
- Average Order Value (AOV)
- Revenue per Customer
- Retention Rate
- Churn Rate


## 6. Success Criteria

### Business Success:
- Increase repeat purchases by 10%
- Reduce churn rate by 8–15%
- Improve targeted campaign efficiency

### Technical Success:
- Meaningful customer segments (Silhouette Score > 0.5)
- Churn model accuracy > 80%
- High Recall for churn prediction


## 7. Stakeholders

- Marketing Team
- Data Analytics Team
- Executive Leadership
- Product Strategy Team


## 8. Constraints

- Limited historical data duration
- No explicit churn label
- Potential missing values
- Highly imbalanced churn cases

## 9. Assumptions

- Past behavior predicts future behavior
- Repeat customers are higher-value
- Segmentation improves personalization


## 10. Risks

- Data leakage during churn modeling
- Poor cluster interpretability
- Overfitting in churn model
- Imbalanced class distribution
